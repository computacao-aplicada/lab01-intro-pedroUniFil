# Exerc√≠cios

1) Exibir apenas o canal vermelho da imagem.

```python
from PIL import Image
import requests
import matplotlib.pyplot as plt
import numpy as np
from io import BytesIO
import random

url = "https://images.unsplash.com/photo-1649930536248-df58fd1f54f8?w=800&auto=format&fit=crop&q=60&ixlib=rb-4.1.0&ixid=M3wxMjA3fDB8MHxzZWFyY2h8N3x8b2JqZXRvcyUyMGFsZWF0JUMzJUIzcmlvc3xlbnwwfHwwfHx8MA%3D%3D"
image = Image.open(requests.get(url, stream=True).raw).convert("RGB")
r, g, b = image.split()

plt.imshow(np.array(r), cmap='Reds')
plt.title("Canal Vermelho (R)")
plt.axis('off')
plt.show()
```

2) Modifique aleatoriamente o canal vermelho e gere a imagem novamente.

```python
r_modified = r.point(lambda x: max(0, min(255, x + random.randint(-50, 50))))
modified_image = Image.merge("RGB", (r_modified, g, b))

plt.imshow(modified_image)
plt.title("Canal Vermelho Modificado Aleat√≥riamente")
plt.axis('off')
plt.show()
```

3) Criar uma imagem negativa (inverter as cores).

```python
np_image = np.array(image)
negative_image = 255 - np_image

plt.imshow(negative_image)
plt.title("Imagem Negativa")
plt.axis('off')
plt.show()
```

4)Salvar uma nova imagem modificada no disco.

```python
negative_pil = Image.fromarray(negative_image.astype('uint8'))
negative_pil.save("imagem_negativa.jpg")
print("Imagem negativa salva como 'imagem_negativa.jpg'")
```

5)Recorte um peda√ßo da imagem e salve em disco.

```python
crop_width = 300
crop_height = 300

width, height = image.size
left = (width - crop_width) // 2
upper = (height - crop_height) // 2
right = left + crop_width
lower = upper + crop_height
cropped_image = image.crop((left, upper, right, lower))
plt.imshow(cropped_image)
plt.title("Recorte")
plt.axis('off')
plt.show()
cropped_image.save("recorte.jpg")
print("Recorte salvo como 'recorte.jpg'")
```
